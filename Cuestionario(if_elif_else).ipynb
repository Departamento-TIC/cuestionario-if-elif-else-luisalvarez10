````python
def quiz():
  questions = [
      {
          "type": "option",
          "question": "¿Cuál es la capital de Francia?",
          "options": ["a) Madrid", "b) París", "c) Londres", "d) Marsella"],
          "answer": "b"
      },
      {
          "type": "number",
          "question": "¿Cuántos años tiene un año bisiesto?",
          "answer": 4
      },
      {
          "type": "option",
          "question": "¿Quién escribió La Odisea?",
          "options": ["a) Góngora", "b) Quevedo", "c) Homero", "d) Benito Pérez Galdós"],
          "answer": "c"
      },
      {
          "type": "number",
          "question": "¿En qué año llegó Cristoba Colón a América?",
          "answer": 1492
      },
      {
          "type": "number",
          "question": "¿Cuántas patas tiene una araña?",
          "answer": 8
      }
  ]

  score = 0

  print("--EMPIEZA EL CUESTIONARIO--\n")

  for q in questions:
    print(q["question"])

    if (q["type"]) == "option":
      for o in q["options"]:
        print(o)



      while True:
        user_answer = input("Introduce tu respuesta (a/b/c/d): ").lower().strip()
        if user_answer in ["a", "b", "c", "d"]:
          if user_answer == q["answer"]:
            print("¡Respuesta correcta!\n")
            score += 1
            break
          else:
            print(f"Respuesta incorrecta. La opción correcta era la {q["answer"]}\n")
            break
        else:
          print("Introduzca una opcion válida")

    else:
      while True:
        try:
          user_answer = int(input("Introduce la respuesta (un número): "))

          if user_answer == q["answer"]:
            print("¡Respuesta correcta!\n")
            score += 1
            break
          else:
            print(f"Respuesta incorrecta. La respuesta era {q["answer"]}\n")
            break
        except ValueError:
          print("ERROR. INTRODUZCA UN NÚMERO.")

  print(f"Has conseguido {score}/{len(questions)} puntos")

quiz()
````
